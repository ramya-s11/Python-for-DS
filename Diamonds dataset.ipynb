```python
diamond = pd.read_csv("diamonds.csv")
diamond
```

11.  Count the duplicate rows of diamonds DataFrame.

```python
duplicate_rows = diamond.duplicated().sum()
duplicate_rows
```

12. Drop rows in case of missing values in carat and cut columns.

```python
diamond.dropna(subset=['carat', 'cut'], inplace=True)
diamond
```

13. Subset the dataframe with only numeric columns.

```python
numeric_col = diamond.select_dtypes(include='number')
numeric_col
```

14. Compute volume as (xyz) when depth is greater than 60. In case of depth less than 60 default volume to 8.

```python
diamond['volume'] = diamond.apply(lambda row: row['x'] * row['y'] * row['z'] if row['depth'] > 60 else 8, axis=1)
diamond
```

15. Impute missing price values with mean.

```python
mean_value = diamond['price'].mean()

# Replace missing values with the mean using fillna()
diamond['price'].fillna(mean_value, inplace=True)
mean_value

diamond.iloc[:10]
```
